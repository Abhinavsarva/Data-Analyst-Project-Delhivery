```python
# Plotting the distribution of 'actual_time'
sns.histplot(dl['actual_time'], kde=True, bins=50, color='gold')
plt.title('Distribution of Actual Time')
plt.xlabel('Actual Time')
plt.ylabel('Frequency')
plt.show()
```

Handling Outliers

```python
Q1 = dl['actual_distance_to_destination'].quantile(0.25)
Q3 = dl['actual_distance_to_destination'].quantile(0.75)
IQR = Q3 - Q1

outliers = dl[(dl['actual_distance_to_destination'] < Q1 - 1.5 * IQR) |
              (dl['actual_distance_to_destination'] > Q3 + 1.5 * IQR)]
outliers
```

Bivariate plot

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(dl, x='actual_distance_to_destination', y='actual_time')
plt.title('Actual Distance vs. Actual Time')
plt.xlabel('Distance (Kms)')
plt.ylabel('Time (Cumulative)')
plt.show()
```
